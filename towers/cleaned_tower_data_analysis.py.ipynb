```python
class copyPoints:
    def __init__(self, point1,point2):
        self.point1=point1;
        self.point2=point2;
    def getp1(self):
        return self.point1
    def getp2(self):
        return self.point2
def gps_dists(point1,point2):
    R = 6373.0 
    dQ = np.radians(point1[0]-point2[0])
    dL = np.radians(point1[1]-point2[1])
    q1 = np.radians(point1[0])
    q2 = np.radians(point2[0])
     
    _a = np.sin(dQ/2)**2+np.cos(q1)*np.cos(q2)*np.sin(dL/2)**2
    _c = 2*np.arctan2(np.sqrt(_a),np.sqrt(1-_a))
    d = _c*R
    return d
def equals(point1,point2, threshold = .150):
    dist = gps_dists(point1,point2)
    if dist<=threshold:
        print dist
        return True
    return False
def copies(points):
    copies = []
    x=0
    for i in range(len(points)-1):
        for j in range(i+1,len(points)-1):
            p1 = points[i]
            p2 = points[j]
            x+=1
                       
            if equals(p1,p2):
                print x
                copy = copyPoints(p1,p2)
                copies.append(copy)
    return copies
def diff(df1,df2):
    merged = df1.merge(df2, indicator=True, how='outer')
    return merged[merged['_merge'] == 'left_only']
    
```

```python
import re
def coordinate_convertor(coordinates):
    lat, lon = re.findall('([0-9]+\-[0-9]+\-[0-9]+\.[0-9]+[SWEN])', coordinates)
    lat_deg,lat_min, lat_sec= lat.split('-')
    lat_sec = max(re.findall('[0-9]+\.[0-9]+', lat_sec), key=len)
    lon_deg,lon_min, lon_sec= lon.split('-')
    lon_sec = max(re.findall('[0-9]+\.[0-9]+', lon_sec), key=len)
    lat_val = float(lat_deg.strip())+float(lat_min.strip())/60.0+float(lat_sec.strip())/3600.0
    lon_val = float(lon_deg.strip())+float(lon_min.strip())/60.0+float(lon_sec.strip())/3600.0
    return lat_val, -lon_val
```

```python
breaks = ['Inc.','LP','Inc.','LLC','LLC','','','','','','','']
def clean(name_parts):
    
    new_parts = []
    for part in name_parts:
        if part.strip()!='':
            new_parts.append(part.strip())
    return new_parts
def descriptive(name):
    name_parts = clean(str(name).split(' '))
    #print 'name_parts: ',name_parts
    for i in range(len(name_parts)):
        part = name_parts[i]
        if len(part)>1:
            part = clean(part.split('/'))[0]
            part = clean(part.split(','))[0]
            name_parts[i]=part
    new_name = ''
    for i in range(len(name_parts)):
        part = name_parts[i]
        if i==len(name_parts)-1 and len(name_parts)>=3:
            if len(part)<=3 or ('.' in part):
                return new_name
        if new_name=='':
            new_name+=part
        else:
            new_name+=' '+part
    return new_name
def nameRemap(names):
    mapper = np.vectorize(descriptive)
    return mapper(names)
        
companies['names'] = nameRemap(names)     
```
